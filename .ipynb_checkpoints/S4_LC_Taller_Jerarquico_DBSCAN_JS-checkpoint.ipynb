### 3. Clustering Jerárquico

#### 3.1. Dendrogramas

Agrupe a los clientes utilizando sólo dos variables: su ingreso y el puntaje de gastos. Para ello retenga estas dos variables, estandarízelas, construya y grafique un dendograma. Para la construcción del dendrograma utilice el enlace 'ward' y la distancia euclidiana. Explique la importancia de estandarizar los datos.

```python
# Retener solo las dos variables: Ingreso y Puntaje_Gasto
variables_seleccionadas = datos_clientes[['Ingreso', 'Puntaje_Gasto']]

# Estandarizar los datos
scaler = StandardScaler()
datos_estandarizados = scaler.fit_transform(variables_seleccionadas)

# linkage utilizando el método 'ward' y distancia euclidiana
linked = linkage(datos_estandarizados, method='ward', metric='euclidean')

# Dendograma
plt.figure(figsize=(12, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='ascending',
           show_leaf_counts=True)
plt.title('Dendograma de Clientes')
plt.xlabel('Índice de Clientes')
plt.ylabel('Distancia Euclídea')
plt.show()
```

Se utiliza la función linkage con el método ward que produce clusters del mismo tamaño y busca minimizar la varianza intra-clusters, y la métrica de distancia Euclídea. La estandarización de los datos es un paso crucial en el análisis de clustering, especialmente cuando se utilizan métodos como el enlace Ward que dependen de la distancia entre los puntos. Es importante porque las variables pueden tener diferentes unidades o escalas (por ejemplo, el ingreso puede estar en miles de dólares, mientras que el puntaje de gasto es una puntuación entre 0 y 100). Si los datos no se estandarizan, las variables con valores numéricos más grandes dominarán el cálculo de las distancias, lo que podría sesgar el análisis y llevar a conclusiones incorrectas. Estandarizar los datos asegura que cada variable contribuya de manera equitativa al proceso de agrupamiento.

### 4. DBSCAN

Implemente DBSCAN, al igual que el punto anterior sólo dos variables: su ingreso y el puntaje de gastos. Al implementar  justifique su elección de `min_samples` y `eps`. ¿Cuántos grupos o clusters encuentra? ¿Cuáles son su características? ¿Cambiaría su estrategia de marketing respecto al punto anterior? Justifique su respuesta.

```python
# Usar NN para estimar un buen valor de eps
neighbors = NearestNeighbors(n_neighbors=8)
neighbors_fit = neighbors.fit(datos_estandarizados)
distances, indices = neighbors_fit.kneighbors(datos_estandarizados)

# Ordenar las distancias para encontrar el "codo" en la curva
distances = np.sort(distances[:, 4])
plt.figure(figsize=(6, 5))
plt.plot(distances)
plt.title('Curva de distancias para estimar eps')
plt.xlabel('Puntos ordenados por distancia')
plt.ylabel('Distancia al 5º vecino más cercano')
plt.show()

i = np.arange(len(distances))
knee = KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')
print(distances[knee.knee])
```

Se utiliza el método de los vecinos más cercanos para seleccionar el parámetro eps. Este método devuelve dos matrices, una que contiene la distancia a los n_neighbors puntos más cercanos y la otra que contiene el índice para cada uno de esos puntos. En la gráfica se puede ver que la parte exponencial de la curva comienza alrededor de una distancia de 0.42. Este valor define el radio máximo de un punto para ser considerado como parte de un cluster.  

Para el parámetro ``min_samples`` se decide usar **2*dim = 4**.

```python
# Configuración final del modelo DBSCAN
eps_value = 0.42
min_samples_value = 4

# Crear y ajustar el modelo DBSCAN
db = DBSCAN(eps=eps_value, min_samples=min_samples_value).fit(datos_estandarizados)
clusters_dbscan = db.labels_

# Contar el número de clusters (excluyendo el ruido)
n_clusters_dbscan = len(set(clusters_dbscan)) - (1 if -1 in clusters_dbscan else 0)

# Crear un DataFrame con los resultados
resultado_dbscan = pd.DataFrame(datos_estandarizados, columns=['Ingreso', 'Puntaje_Gasto']).assign(Cluster=clusters_dbscan)

# Generar el gráfico
plt.figure(figsize=(7, 5))
for k in np.unique(clusters_dbscan):
    if k != -1:
        plt.scatter(
            x=resultado_dbscan.loc[resultado_dbscan.Cluster == k, 'Ingreso'], 
            y=resultado_dbscan.loc[resultado_dbscan.Cluster == k, 'Puntaje_Gasto'],
            label=f'Cluster {k}'
            )
    else:
        plt.scatter(
            x=resultado_dbscan.loc[resultado_dbscan.Cluster == k, 'Ingreso'], 
            y=resultado_dbscan.loc[resultado_dbscan.Cluster == k, 'Puntaje_Gasto'],
            c="black",
            label='Outliers (ruido)'
            )
plt.title(f'Se encontraron {n_clusters_dbscan} clusteres para epsilon {eps_value}')
plt.xlabel('Ingreso (Estandarizado)')
plt.ylabel('Puntaje de Gasto (Estandarizado)')
plt.legend()
plt.show()
```

Los clusters identificados pueden ser útiles para detectar segmentos de clientes específicos, incluyendo aquellos que no siguen los patrones generales (ruido), que podrían necesitar estrategias de marketing especializadas. Al comparar con el método Jerárquico Aglomerativo, DBSCAN puede identificar grupos de densidad variable y manejar en parte el ruido, pero el método jerárquico aglomerativo proporciona una segmentación más clara y estructurada para aplicaciones de marketing. Dependiendo de los resultados.  

### 5. Incorporando más Dimensiones

Hasta este momento hemos utilizado solo dos variables para nuestro análisis, sin embargo, la base cuenta con otras variables más: género y edad. Usando las categorías de edad creadas anteriormente incorpore género y grupo etario a su análisis. Genere los grupos utilizando DBSCAN y la distancia de Gower, justificando su elección de `min_samples` y `eps`   ¿Incorporar estas variables, ayuda a la focalización de la estrategia de marketing?

```python
# Preparación de datos
datos_clientes['Genero'] = datos_clientes['Genero'].astype('category').cat.codes
datos_clientes['Grupo_Etario'] = datos_clientes['Grupo_Etario'].astype('category').cat.codes

# Seleccionar las variables para el análisis
variables = ['Ingreso', 'Puntaje_Gasto', 'Genero', 'Grupo_Etario']
datos_seleccionados_gower = datos_clientes[variables]

# Calcular la distancia de Gower
gower_distances = gower_matrix(datos_seleccionados_gower)

# Usar el método de los vecinos más cercanos para estimar un buen valor de eps
neighbors = NearestNeighbors(n_neighbors=8, metric='precomputed')
neighbors_fit = neighbors.fit(gower_distances)
distances, indices = neighbors_fit.kneighbors(gower_distances)

# Ordenar las distancias para encontrar el "codo" en la curva
distances = np.sort(distances[:, 7])  # min_samples=8, se selecciona el 8º vecino
plt.figure(figsize=(6, 5))
plt.plot(distances)
plt.title('Curva de distancias para estimar eps')
plt.xlabel('Puntos ordenados por distancia')
plt.ylabel('Distancia al 8º vecino más cercano')
plt.show()

# Encontrar el valor preciso del codo
i = np.arange(len(distances))
knee = KneeLocator(i, distances, S=1, curve='convex', direction='increasing', interp_method='polynomial')
optimal_eps = distances[knee.knee]

optimal_eps
```

```python
# Crear el modelo DBSCAN
db = DBSCAN(eps=0.104, min_samples=8, metric='precomputed')
clusters_dbscan_gower = db.fit_predict(gower_distances)

# Contar el número de clusters (excluyendo el ruido)
n_clusters_dbscan_gower = len(set(clusters_dbscan_gower)) - (1 if -1 in clusters_dbscan_gower else 0)

# DataFrame con los resultados
resultado_dbscan_gower = datos_seleccionados_gower.assign(Cluster=clusters_dbscan_gower)

# Gráfico
plt.figure(figsize=(7, 5))
for k in np.unique(clusters_dbscan_gower):
    if k != -1:
        plt.scatter(
            x=resultado_dbscan_gower.loc[resultado_dbscan_gower.Cluster == k, 'Ingreso'], 
            y=resultado_dbscan_gower.loc[resultado_dbscan_gower.Cluster == k, 'Puntaje_Gasto'],
            label=f'Cluster {k}'
            )
    else:
        plt.scatter(
            x=resultado_dbscan_gower.loc[resultado_dbscan_gower.Cluster == k, 'Ingreso'], 
            y=resultado_dbscan_gower.loc[resultado_dbscan_gower.Cluster == k, 'Puntaje_Gasto'],
            c="black",
            label='Ruido'
            )
plt.title(f'Se encontraron {n_clusters_dbscan_gower} clusteres para epsilon {0.104}')
plt.xlabel('Ingreso (Estandarizado)')
plt.ylabel('Puntaje de Gasto (Estandarizado)')
plt.legend()
plt.show()

# Descripción de los clusters encontrados (excluyendo el ruido)
cluster_descripcion_dbscan_gower = resultado_dbscan_gower[clusters_dbscan_gower != -1].groupby('Cluster')[['Ingreso', 'Puntaje_Gasto']].mean().reset_index()

n_clusters_dbscan_gower, cluster_descripcion_dbscan_gower
```

Después de optimizar eps en 0.104 y utilizar min_samples = 8 (2*Dim), e incorporar variables adicionales como género y grupo etario, junto con la distancia de Gower, no parece haber proporcionado una segmentación más clara y útil. Solo dos cluster dueron identificados, y esta segmentación sigue siendo muy general y no focaliza muy bien la estrategía de marketing a diferencia del algoritmo jerárquico aglomerativo que si logra diferenciar con más detalle los segmentos de clientes.
